# Regression

* One of the most important statistical and machine learning tools.  
* Allows us to make predictions based upon data by learning the relationship between input and output variables.  
* The output variable dependent on the input variables are continuous valued real numbers.  
* Regression hellps us to understand how the value of output variable changes with respect to the changes in the input variable.  
* Regression techniques are used for the prediction of prices, Economics and varations.  


If my Output is binary categorical like yes/no etc, then it is called as a **Classification Problem.**  

Whenever my Ouput varies in a continuous manner in accordance with Input, then it is called as a **Regression Problem.**

# Simple Linear Regression

* It is the simplest form of linear regression used when there is one single input variable(Input) for the output variable(Target).  
* The input variable(referred to as X) helps in predicting the value of the output variable. 
* The output or target is the variable that we want to predict(y).  

Equation: **ŷ = β0 + β1X + ε**  
* ŷ is target  
* β0, β1 are coefficients  
* X -> input  
* ε -> random error  

* β0, called the intercept, shows the point where the estimated regression line cuts the y-axis.  
* β1, determines the slope of the estimated regression line.  


* The main aim of regression is to, estimate the parameters β0 and β1 from the sample.  
* Once we find the optimum values for these two parameters, a line of best fit can be used to find the values of y given the values of X.  
* The line is used to predict the output of unseen inputs.  



* β0 and β1 values are estimated using Ordinary Least Squares(OLS) method.  
* The main goal is to minimize the dist. from the black dots to the red line as close to zero as possible.  
* It is done by minimizing the squared distances between actual and predicted outcomes.  
* The difference between actual and predicted value is called the residual(e).  
* It can be positive/negative depending on whether the model overpredicted or underpredicted the outcome.  
* To find all the residual error, take the sum of squares of error terms and it is called the Residual Sum of Squares(RSS)  

 - subtracting, actual value with the predicted value and squaring them.  
 

* The OLS method reduces the residual sum of squares(RSS).


**Advantage:** Easy to implement, performs well for linearly seperable data.  
**Disadvantage:** It is linear, but data may or maynot be. It is prone to noise and overfitting.

### Different Kinds of Relationship

* **Positive Relationship:** When the regression line between two variables moves in the same direction with an upward slope, then the variables are said to be positively correlated.  
* If we increase the value of x(Independent variable), then we will see an increase in the dependent variable(y).  


* **Negative Relationship:** When the regression line between two variables moves in the same direction with a downward slope, then the variables are said to be in a negative relationship.  
* If we increase the value of x(Independent variable), then we will see an decrease in the dependent variable(y).

# Employee Salary Prediction

```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Importing Data
data = pd.read_csv('Datasets/Salary.csv')
```

```python
data
```

```python
data.shape
```

```python
# Checking Dataset information
data.info()
```

```python
# Checking for none values
data.isnull().count()
```

```python
# Model Building:
# Splitting the Dataset into independent and dependent variables

X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
# independent variable
X
```

```python
# dependent variable
y
```

```python
# Scatter plot to check relationship between independent variable and dependent variable
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue')
```

```python
# Splitting dataset for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Displaying y_test
y_test
```

```python
# Fitting Simple Linear Regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
# Calculating error/residue
residue = y_pred - y_test
residue
```

```python
# Visualization:
# Visualizing the training set results
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="yellow")
plt.title("Salary Vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing the test set results
plt.scatter(X_test, y_test, color="red")
plt.plot(X_train, regressor.predict(X_train), color="green")
plt.title("Salary Vs Experience(Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Calculating the intercept and coefficient
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE =", rmse) # Root Mean Square Error (RMSE) is the standard deviation of the residuals(prediction errors)
print("R2 Score =",r2) # Accuracy score
```
