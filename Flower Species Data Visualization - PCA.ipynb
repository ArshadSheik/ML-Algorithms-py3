# Principal Component Analysis (PCA)

* It is an Unsupervised Machine Learning algorithm that is used for dimensionality reduction in Machine Learning.  
* It is a statistical process that converts the observations of correlated features into a set of linearly uncorrelated features.  
* This process is done with the help of some orthogonal Transformation.  
* The new transformated features are called as Principal components.  
* It is one of the common tools that is used for EDA and predictive modelling.  
* This technique is used to draw some patterns from the dataset by reducing the variances.  
* It tries to find the low - dimensional surfacce in order to project the high dimensional data.  

* This technique works by considering the variance of each attribute because the high attribute shows the good split  between the classes.  
* PCA reduces the dimensionality.  
* PCA techniques are employed in Image processing, Movie Recommendation System, and it is used to optimize the power allocation in various communication channels.  
* It is a type of feature extraction technique, it contains the important variables and it drops the less important variables.

### In PCA, only Most important features are selected.
* PCA is based on the mathematical concepts of:
    - Variance and Covariance.  
    - Eigen values and Eigen factors.  

* Common terms in PCA algorithm:

    - **Dimensionality:** It is the total number of variables present in the given dataset. (No. of cols present in dataset)  
    
    - **Correlation:** It defines the relationship between any two variables. If one variable chages, the other also gets changed.  
        - correlation value ranges from -1 to +1  
        - -1 indicates variables are inversely proportional to each other and +1 indicates variables are directly proportional to each other.  
        
    - **Orthogonal:** It defines that the variables are not related to each other and hence the correlation between the pair of variables is 0  
    
    - **Eigen Vectors:** If there is a square matrix M and a non-zero vector v is given, V will be the eigen vector if Av is scalar multiple of v  
    
    - **Covariance Matrix:** The matrix containing the covariance between a pair of variables is called the covariance matrix.

## PCA

* The transformed new features or the output of PCA are the Principal components.  
* The No. of these Principal components are either equal to or less than the original features present in the dataset.  

### Properties of  Principal components:

* The Principal component must be the linear combination of the original features.  
* All the components are orthogonal, i.e., the correlation between a pair of variables is zero.  
* The importance of each component decreases when going from 1 to n.  
* It means that 1PC has the most importance and nPC will have the least importance.   

## Steps for PCA Algorithms:

### -> Dataset Collection  

* Input data is collected and divided into two subparts X and Y.  
* X is training set, Y is Validation set  

### -> Representing Data into A Structure  

* Dataset is organized into a structure, as 2-D Matrix of independent variable X.  
* Each row corresponds to data items and the columns denotes the features.  
* No. of cols is the dimensions of the dataset.  

### -> Standardizing the Data  

* Data Standardization is performed.  
* In a specific column, features with high variance are more important when compared to the features with low variance.  
* If the importance of features is independent of the variance of the features, then each data item in the column is divided with the standard deviation of the column.  
* The matrix is named as Z.  

### -> Calculating the Covariance of Z

* To calculate covariance of Z, matrix Z is taken and the transpose of matrix is taken.  
* After transpose, the matrix is multiplied by Z.  
* The output matrix will be the covariance matrix Z  

### -> Calculating Eigen values & Eigen vectors  

* We need to calculate Eigen values & Eigen vectors for the resultant covariance matrix Z.  
* Eigen vectors are the directions of the axes with high information.  
* The coefficients of the Eigen vectors are defined as Eigen values.  

### -> Sorting the Eigen vectors

* Eigen values are taken and sorted in the descending order from largest to smallest.  
* Eigen vectors are sorted accordingly in matrix P of eigen values.  
* The resultant matrix is named as P*  

### -> Compute the new features or the Principal components  

* New features are calculated  
* In order to achieve this, P* matrix is multiplied with matrix Z  
* In the resultant matrix Z*, each observation is the linear combination of the original features.  
* Each column of Z* matrix is independent of each other.  

### -> Eliminate Unessential Features from the new Dataset  
* New feature set has been created  
* Only essential features are kept in the new dataset and the unwanted features are removed.  

## Applications of PCA

* Used as Dimensionality reduction technique in various AI applications.  
* Used in applications like computer vision & Image Compression applications.  
* Used for finding hidden patterns if the data has higher dimensions.  
* Used in some of the fields like Finance, Data Mining, Psychology etc.

## Advantages of PCA  

* Easy computation - based on linear algebra  
* Speeds up other ML Algorithms  
* Counteracts the issue of High Dimensional Data  



## Disadvantages  

* Low Interpretability of Principal Components  
* Trade - Off between information loss and dimensionality reduction

# Flower Species Data Visualization - PCA

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'target'])
```

```python
df.head()
```

```python
# Standardizing the Data
from sklearn.preprocessing import StandardScaler
features = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Separating out the features
X = df.loc[:, features].values

# Separating out the target
y = df.loc[:, ['target']].values

# Standardizing the features
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
# PCA Projection to 2D:
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, 
                          columns=['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df[['target']]], axis=1)
finalDf
```

```python
import matplotlib.pyplot as plt
# Visualize 2D Projection:
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Principal Component 1', fontsize=15)
ax.set_ylabel('Principal Component 2', fontsize=15)
ax.set_title('2 component PCA', fontsize=20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1'], 
               finalDf.loc[indicesToKeep, 'principal component 2'], 
               c = color, s = 50)
    
ax.legend(targets)
ax.grid()
```
